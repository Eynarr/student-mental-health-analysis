# Student Mental Health Analysis

This notebook explores how mental health changes based on
years in the country and social factors.

```python
import pandas as pd

df = pd.read_csv("../data/processed/students.csv")
df.head()
```

### Depression Score by Years in the Country
This plot shows the average depression score for international students based on the number of years in the country.

```python
import matplotlib.pyplot as plt

intl = df[df["student_type"] == "International"]

trend = (
    intl.groupby("years_country")["depression_score"]
    .mean()
    .sort_index()
)

trend.plot()
plt.title("Average Depression Score by Years in Country")
plt.xlabel("Years in Country")
plt.ylabel("Depression Score")
plt.show()
```

There is no clear increasing or decreasing trend. 
Most data is concentrated in the first years, so later years may not be reliable.

### Social Connectedness vs Depression
This plot shows the relationship between social connectedness and depression.

```python
group_avg = df.groupby("student_type")["depression_score"].mean()

group_avg.plot(kind="bar")
plt.title("Average Depression by Student Type")
plt.ylabel("Depression Score")
plt.show()
```

Students with lower social connectedness tend to have higher depression scores.

### Average Depression by Student Type
This plot compares the average depression score between student types.

```python
group_avg = df.groupby("student_type")["depression_score"].mean()

group_avg.plot(kind="bar")
plt.title("Average Depression by Student Type")
plt.ylabel("Depression Score")
plt.show()
```

There is a small difference between groups, but it is not very large.

### Conclusions

- Most students are in early years (1–4)
- There is no clear trend of depression over time
- Social connectedness seems related to depression
- Both domestic and international students can have high depression scores
